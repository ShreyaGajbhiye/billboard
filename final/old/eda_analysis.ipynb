```python
import requests
import base64
import json

# Spotify credentials
client_id = '45061234f7294375a7dff6300de09df2'
client_secret = '01333f4747234a28a683f37daf930cd2'
```

```python


def get_access_token(client_id, client_secret):
    token_url = 'https://accounts.spotify.com/api/token'
    client_creds_b64 = base64.b64encode(f"{client_id}:{client_secret}".encode())
    token_headers = {"Authorization": f"Basic {client_creds_b64.decode()}"}
    token_data = {"grant_type": "client_credentials"}
    r = requests.post(token_url, headers=token_headers, data=token_data)
    return r.json()['access_token']

def search_track(track_name, access_token):
    search_url = 'https://api.spotify.com/v1/search'
    headers = {"Authorization": f"Bearer {access_token}"}
    params = {"q": track_name, "type": "track", "limit": 1}
    r = requests.get(search_url, headers=headers, params=params)
    results = r.json()
    tracks = results['tracks']['items']
    if tracks:
        return tracks[0]['id']
    else:
        return None

def get_track_release_date(track_id, access_token):
    track_url = f"https://api.spotify.com/v1/tracks/{track_id}"
    headers = {"Authorization": f"Bearer {access_token}"}
    r = requests.get(track_url, headers=headers)
    track_data = r.json()
    return track_data['album']['release_date']





```

```python
def get_date(track_name):
    print(track_name)
    access_token = get_access_token(client_id, client_secret)
    track_id = search_track(track_name, access_token)

    if track_id:
        release_date = get_track_release_date(track_id, access_token)
        return pd.to_datetime(release_date).year
    else:
        return np.nan
```

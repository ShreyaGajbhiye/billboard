# Correlations between Song Characteristics

```python
corr_df = bb3.drop(["Year", "Name", "Artist", "Genre"], axis = 1).corr(method = "pearson")

plt.figure(figsize = (15,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths = 1, linecolor="Black")
heatmap.set_title("Corrrelation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

Danceability and Valence (0.4 correlation coefficient) - a song being more positive (cheerful, happy) is positively correlated to the song being easier to dance to.


Energy and Valence (0.4 correlation coefficient) - a song being more positive (cheerful, happy) is positively correlated to the song being more energetic.

Energy and Loudness (0.7 correlation coefficient) - a song having more energy is positively correlated to the song being more loud.

Explicit and Speechiness (0.5 correlation coefficient) - a song having more number of spoken words generally express more stronger language/matured theme. 

Explicit and Danceability (0.3 correlation coefficient)  - A song having presence of stronger language or a matured theme are easier to dance to 

Loudness and Valence (0.3 correlation coefficient) - a song is more positive (cheerful, happy) when it is more loud. When a song is more quiet, it is to some extent, less cheerful

Speechiness and Danceability (0.2 correlation coefficient) - a song having more spoken words is positively correlated to the song being easier to dance to. 

Valence and Duration (-0.2 correlation coefficient) - a song being longer is resulting in the song being less positive (cheerful, happy). We can see that more clearly when we look at it from the genre point of view.

Tempo and Danceability (-0.3 correlation coefficient) - A fast paced song is generally a less danceable song. 

Loudness and Acousticness (-0.4 correlation coefficient) - a song is louder when it is less acoustic. That is expected because usually acoustic songs are more calm/quiet than electronic music.

Energy and Duration (-0.2 correlation coefficient) - a song having more energy is generally less longer.

Energy and Acousticness (-0.5 correlation coefficient) - a song being more acoustic has a negative correlation with its energy, the more the acousticness the less the energy.

Valence and Acousticness (-0.2 correlation coefficient) - a song being more positive (cheerful, happy) is negatively correlation to the song being acoustic.

```python
recent0_year_bb3 = bb3[bb3.Year == 2017].mean()[1:-1]
recent1_year_bb3 = bb3[bb3.Year == 2018].mean()[1:-1]
recent2_year_bb3 = bb3[bb3.Year == 2019].mean()[1:-1]
```

```python
data = {"Averages in 2017": recent0_year_bb3,
        "Averages in 2018": recent1_year_bb3,
        "Averages in 2019": recent2_year_bb3}

forecasting_bb3 = pd.DataFrame(data)

forecasting_bb3
```

Now, and using the averages above, we are going to calculate the percentage of change these characteristics are witnessing over those 3 years. From those percentages of change between the years, we will be calculating the average percentage of change and using it for future predictions and forecasting.

```python
forecasting_bb3["Percentage of Change 2017-2018"] = np.abs(forecasting_bb3["Averages in 2018"] - forecasting_bb3["Averages in 2017"]) / forecasting_bb3["Averages in 2017"] *100
```

```python
forecasting_bb3["Percentage of Change 2018-2019"] = np.abs(forecasting_bb3["Averages in 2019"] - forecasting_bb3["Averages in 2018"]) / forecasting_bb3["Averages in 2018"] *100
```

```python
forecasting_bb3
```

```python
forecasting_bb3["Average Percentage of Change"] = forecasting_bb3[["Percentage of Change 2017-2018", "Percentage of Change 2018-2019"]].mean(axis=1)
```

```python
forecasting_bb3
```

```python
forecasting_bb3["Averages in 2020"] = 0
forecasting_bb3["Averages in 2021"] = 0
forecasting_bb3["Averages in 2022"] = 0
forecasting_bb3["Averages in 2023"] = 0

forecasting_bb3.iloc[:,5:]
```

```python
features = ["Loudness", "Energy", "Danceability", "Tempo", "Liveness", "Valence", "Duration", "Acousticness", "Speechiness", "Instrumentalness", "Explicit"]
for feat in features:
    if forecasting_bb3["Averages in 2019"].loc[feat] > forecasting_bb3["Averages in 2018"].loc[feat]:
        forecasting_bb3["Averages in 2020"].loc[feat] = forecasting_bb3["Averages in 2019"].loc[feat] + (forecasting_bb3["Averages in 2019"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    else:
        forecasting_bb3["Averages in 2020"].loc[feat] = forecasting_bb3["Averages in 2019"].loc[feat] - (forecasting_bb3["Averages in 2019"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    if forecasting_bb3["Averages in 2020"].loc[feat] > forecasting_bb3["Averages in 2019"].loc[feat]:
        forecasting_bb3["Averages in 2021"].loc[feat] = forecasting_bb3["Averages in 2020"].loc[feat] + (forecasting_bb3["Averages in 2020"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    else:
        forecasting_bb3["Averages in 2021"].loc[feat] = forecasting_bb3["Averages in 2020"].loc[feat] - (forecasting_bb3["Averages in 2020"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    if forecasting_bb3["Averages in 2021"].loc[feat] > forecasting_bb3["Averages in 2020"].loc[feat]:
        forecasting_bb3["Averages in 2022"].loc[feat] = forecasting_bb3["Averages in 2021"].loc[feat] + (forecasting_bb3["Averages in 2021"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    else:
        forecasting_bb3["Averages in 2022"].loc[feat] = forecasting_bb3["Averages in 2021"].loc[feat] - (forecasting_bb3["Averages in 2021"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    if forecasting_bb3["Averages in 2022"].loc[feat] > forecasting_bb3["Averages in 2021"].loc[feat]:
        forecasting_bb3["Averages in 2023"].loc[feat] = forecasting_bb3["Averages in 2022"].loc[feat] + (forecasting_bb3["Averages in 2022"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
    else:
        forecasting_bb3["Averages in 2023"].loc[feat] = forecasting_bb3["Averages in 2022"].loc[feat] - (forecasting_bb3["Averages in 2022"].loc[feat] * (forecasting_bb3["Average Percentage of Change"].loc[feat] / 100))
```

```python
temp_df = pd.DataFrame(bb3["Genre"].value_counts())
temp_df
```

```python
country_count = 0
other_count = 0
rap_count = 0
pop_count = 0
rock_count = 0
for genre in temp_df.index:
    if "Country" in genre:
        country_count += temp_df["Genre"].loc[genre]
    elif "Rock" in genre:
        rap_count += temp_df["Genre"].loc[genre]
    elif "Rap" in genre:
        rock_count += temp_df["Genre"].loc[genre]
    elif "Pop" in genre:
        pop_count += temp_df["Genre"].loc[genre]
    else:
        other_count += temp_df["Genre"].loc[genre]
```

```python
print("The amount of songs from the Country genre on the billboard: ", country_count, "Songs")
print("The amount of songs from the Rap genre on the billboard: ", rap_count, "Songs")
print("The amount of songs from the Rock genre on the billboard: ", rock_count, "Songs")
print("The amount of songs from the Pop genre on the billboard: ", pop_count, "Songs")
print("The amount of songs from the remaining genres on the billboard: ", other_count, "Songs")
print("The total amount of songs on the billboard: ", temp_df.Genre.sum(), "Songs")
```

```python
percentage_pop = (pop_count * 100) / temp_df.Genre.sum()
percentage_rap = (rap_count * 100) / temp_df.Genre.sum()
percentage_country = (country_count * 100) / temp_df.Genre.sum()
percentage_rock = (rock_count * 100) / temp_df.Genre.sum()

percentage_other = (other_count * 100) / temp_df.Genre.sum()
```

```python
data = [percentage_pop, percentage_rap,percentage_country,percentage_rock, percentage_other]
keys = ["Pop Music", "Rap Music", "Country Music", "Rock Music", "Other Genres"]

# explode = [0.1, 0.1,0.1,0.1, 0]

plt.figure(figsize=(15,6))

plot12 = plt.pie(data, labels = keys, autopct='%.0f%%')
plt.title("Music Genres on the Billboard")
plt.show()
```
